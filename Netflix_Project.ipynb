### Data Cleaning

#### Removing Duplicate Rows (Considering all columns)


```python
df.drop_duplicates(keep='last',inplace=True)
```

```python
df.size
```

#### Handling Null Values

```python
df.isnull().sum()
```

```python
df[df.isnull()]
```

```python
df['director'].fillna('Not Specified',inplace=True)
```

```python
df['cast'].fillna('Not Specified',inplace=True)
```

```python
df['country'].fillna('Unknown',inplace=True)
```

```python
df['rating'].fillna('Not Rated',inplace=True)
```

```python
# Removing extra Spaces
df['type'].str.strip()
df['title'] = df['title'].str.strip()
df['country'] = df['country'].str.strip()
df['rating'] = df['rating'].str.strip()
```

```python
df.isnull().sum()
```

#### Cleaning Not Null Values

```python
df['type'].value_counts()
```

```python
df['release_year'].value_counts()
```

```python
df['rating'].value_counts()
```

```python
## some are in duration format 
df['rating'] = df['rating'].replace({
    '84 min': 'Not Rated',
    '66 min': 'Not Rated',
    '74 min': 'Not Rated'})
```

```python
df['rating'].value_counts()
```

```python
df['duration'].value_counts()
```

#### Feature engineering

```python
# df['date_added']	Selects the column date_added in your DataFrame
#pd.to_datetime()	Converts values to datetime64 format
#errors='coerce'	If a value can’t be converted (e.g., it's missing or invalid), it becomes NaT (Not a Time — like NaN for dates)
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

missing_date_mask = df['date_added'].isnull()
# Fill missing date_added using December 1st of release year
df.loc[missing_date_mask, 'date_added'] = pd.to_datetime(
    df.loc[missing_date_mask, 'release_year'].astype(str) + "-12-01"
)
```

```python
df['duration'].fillna('Unknown',inplace=True)
```

```python
df.isnull().sum()  
# All null values are handled
```

```python
df.head()
```

```python
# Extracting year,month from date_added
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month
```

```python
# Main Genre
df['main_genre'] = df['listed_in'].str.split(',').str[0].str.strip()
```

```python
df.head(5)
```

```python
# Duration Breaking
#str.extract(...)	Extracts part of the string using a regular expression
#r'([a-zA-Z]+)'	This regex grabs all letters (the word part),expand=False:Returns a Series instead of a DataFrame
df['duration_type'] = df['duration'].str.extract(r'([a-zA-Z]+)', expand=False)
```

```python
#r'(\d+)'	This regex captures the digits only (e.g., 90, 2, 1)
#.astype(float)	Converts the extracted string number to numeric (float) for analysis
df['duration_value'] = df['duration'].str.extract(r'(\d+)', expand=False).astype(float)
```
